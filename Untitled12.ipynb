```python
# Node untuk Linked List
class Node:
    def __init__(self, plat):
        self.plat = plat
        self.next = None


# Queue menggunakan Linked List
class QueueParkir:
    def __init__(self, kapasitas):
        self.front = None
        self.rear = None
        self.kapasitas = kapasitas
        self.size = 0

    # Cek kosong
    def is_empty(self):
        return self.front is None

    # Cek penuh
    def is_full(self):
        return self.size == self.kapasitas

    # Tambah mobil (enqueue)
    def masuk(self, plat):
        if self.is_full():
            print("Parkiran penuh!")
            return

        new_node = Node(plat)

        if self.is_empty():
            self.front = self.rear = new_node
        else:
            self.rear.next = new_node
            self.rear = new_node

        self.size += 1
        print(f"Mobil {plat} masuk parkiran.")

    # Keluar mobil (dequeue)
    def keluar(self):
        if self.is_empty():
            print("Parkiran kosong!")
            return

        keluar = self.front
        self.front = self.front.next

        if self.front is None:
            self.rear = None

        self.size -= 1
        print(f"Mobil {keluar.plat} keluar parkiran.")

    # Tampilkan isi parkiran
    def tampilkan(self):
        if self.is_empty():
            print("Parkiran kosong.")
            return

        current = self.front
        print("Isi parkiran:")
        while current:
            print(f"- {current.plat}")
            current = current.next


# Program utama
if __name__ == "__main__":
    parkir = QueueParkir(5)

    while True:
        print("\n=== SISTEM PARKIR ===")
        print("1. Mobil Masuk")
        print("2. Mobil Keluar")
        print("3. Lihat Parkiran")
        print("4. Keluar")

        pilihan = input("Pilih menu: ")

        if pilihan == "1":
            plat = input("Masukkan nomor plat: ")
            parkir.masuk(plat)

        elif pilihan == "2":
            parkir.keluar()

        elif pilihan == "3":
            parkir.tampilkan()

        elif pilihan == "4":
            print("Terima kasih!")
            break

        else:
            print("Pilihan tidak valid!")
```
